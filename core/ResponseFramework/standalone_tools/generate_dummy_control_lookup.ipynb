```python
import pandas as pd
import numpy as np
import itertools

# ==== USER INPUTS ====
n_turbines = 5

wsp_bins = np.array([0, 5, 10, 15, 20, 25])     # m/s
wdir_bins = np.array([0, 90, 180, 270, 360])    # degrees
TI_bins = np.array([0.05, 0.10, 0.15, 0.20, 0.25, 0.30])

# ==== BIN CENTERS ====
def bin_centers(bins):
    return [(bins[i] + bins[i+1]) / 2 for i in range(len(bins)-1)]

wsp_centers = bin_centers(wsp_bins)
wdir_centers = bin_centers(wdir_bins)
TI_centers = bin_centers(TI_bins)

# ==== GENERATE ALL COMBINATIONS ====
rows = []

for wsp_bin, wdir_bin, TI_bin in itertools.product(wsp_centers, wdir_centers, TI_centers):
    row = {
        "wsp_bin": wsp_bin,
        "wdir_bin": wdir_bin,
        "TI_bin": TI_bin
    }
    # Add yaw and power for each turbine
    for tid in range(n_turbines):
        row[f"yaw_{tid}"] = np.random.uniform(-30, 30)    # deg
        row[f"power_{tid}"] = 100# np.random.uniform(0.5, 1.0) # normalized
    rows.append(row)

lookup_df = pd.DataFrame(rows)

# ==== SAVE ====
output_file = "data/controller_inputs/dummy_lookup_controller.csv"
lookup_df.to_csv(output_file, index=False)

print(f"Generated {len(lookup_df)} entries for {n_turbines} turbines.")
print(f"Saved to {output_file}.")
```
