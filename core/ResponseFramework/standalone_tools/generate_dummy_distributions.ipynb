```python
import numpy as np
import pandas as pd

# 1. Define ranges and bin widths
wsp_min, wsp_max, wsp_bin = 3, 25, 2       # Wind speed: from 4 to 12 m/s, step 2
TI_min, TI_max, TI_bin = 0.05, 0.25, 0.04  # Turbulence Intensity: from 0.08 to 0.16, step 0.02
wdir_min, wdir_max, wdir_bin = 0, 360, 30  # Wind direction: from 0 to 360 deg, step 90
price_min =0
price_max = 120



# 2. Create bin centers and round to 2 decimals
wsp_bins = np.arange(wsp_min, wsp_max + wsp_bin, wsp_bin)
wsp_centers = np.round((wsp_bins[:-1] + wsp_bins[1:]) / 2, 2)

TI_bins = np.arange(TI_min, TI_max + TI_bin, TI_bin)
TI_centers = np.round((TI_bins[:-1] + TI_bins[1:]) / 2, 2)

wdir_bins = np.arange(wdir_min, wdir_max, wdir_bin)
wdir_centers = np.round((wdir_bins[:-1] + wdir_bins[1:]) / 2, 2)

# 3. Create all combinations of wsp, TI, wdir
combinations = [(w, t, d) for w in wsp_centers for t in TI_centers for d in wdir_centers]

# 4. Assign random probabilities that sum to 1 (no rounding)
random_probs = np.random.rand(len(combinations))
random_probs /= random_probs.sum()


# 5. Create DataFrame
df = pd.DataFrame(combinations, columns=['wsp', 'TI', 'wdir'])
df['probability'] = random_probs
# 5.2 Generate random prices in the user-defined range
df['price'] = np.round(np.random.uniform(price_min, price_max, size=len(df)), 2)

# 6. Save to CSV
df.to_csv('distributions/joint_3D_distribution_long.csv', index=False)

df.head()
```
