```python
import numpy as np
import pandas as pd
from pathlib import Path

# Output folder
output_dir = Path("data/timeseries")
output_dir.mkdir(exist_ok=True)

# Settings
start = "2021-01-01 00:00:00"
periods = 168  # One week of hourly data
freq = "10min"

# Generate timestamps
timestamps = pd.date_range(start=start, periods=periods, freq=freq)

# --- Wind Time Series ---
wind_df = pd.DataFrame({
    "timestamp": timestamps,
    "wsp": np.random.uniform(2, 14, size=periods),  # wind speed in m/s
    "TI": np.random.uniform(0.02, 0.24, size=periods),  # turbulence intensity
    "wdir": np.random.uniform(0, 360, size=periods)  # wind direction
})
# wind_df.to_csv(output_dir / "wind_timeseries_1hr.csv", index=False)

# --- Price Time Series ---
price_df = pd.DataFrame({
    "timestamp": timestamps,
    "price": np.random.uniform(20, 80, size=periods)  # price in EUR/MWh
})
# price_df.to_csv(output_dir / "price_timeseries_1hr.csv", index=False)

# Merge on timestamp
combined_df = pd.merge(wind_df, price_df, on="timestamp")

# Save single CSV
combined_df.to_csv(output_dir / "wind_price_timeseries_10min.csv", index=False)

print("Combined wind and price timeseries saved as csv in ./data/timeseries/")
```
